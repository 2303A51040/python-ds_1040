```python
#1
import pandas as pd

# Create two sample datasets
students = pd.DataFrame({
    'student_id': [1, 2, 3, 4, 5],
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'class': ['A', 'A', 'B', 'B', 'C']
})

scores = pd.DataFrame({
    'student_id': [1, 2, 3, 5, 6],
    'exam_score': [85, 90, 78, 92, 88]
})

print("Students Dataset:\n", students, "\n")
print("Scores Dataset:\n", scores, "\n")

# --- Merging ---
inner_merge = pd.merge(students, scores, on='student_id', how='inner')
left_merge = pd.merge(students, scores, on='student_id', how='left')
right_merge = pd.merge(students, scores, on='student_id', how='right')

print("Inner Merge:\n", inner_merge, "\n")
print("Left Merge:\n", left_merge, "\n")
print("Right Merge:\n", right_merge, "\n")

# --- Aggregate Statistics ---
print("Aggregate Statistics:")
print("Mean Score:", inner_merge['exam_score'].mean())
print("Max Score:", inner_merge['exam_score'].max())
print("Min Score:", inner_merge['exam_score'].min())
print("Count of Scores:", inner_merge['exam_score'].count())

# --- Group By Example (class-wise average score) ---
group_stats = inner_merge.groupby('class')['exam_score'].agg(['mean', 'max', 'min', 'count'])
print("\nClass-wise Statistics:\n", group_stats)
```

```python
# 2
customers = pd.DataFrame({
    'customer_id': [101, 102, 103, 104, 105],
    'customer_name': ['John', 'Sara', 'Mike', 'Linda', 'Tom'],
    'region': ['North', 'South', 'East', 'West', 'North']
})

sales = pd.DataFrame({
    'sale_id': [1, 2, 3, 4, 5, 6, 7],
    'customer_id': [101, 102, 101, 103, 105, 105, 106],
    'amount': [200, 150, 300, 400, 250, 500, 100]
})

print("Customers Dataset:\n", customers, "\n")
print("Sales Dataset:\n", sales, "\n")

# --- Merge ---
inner_merge_sales = pd.merge(customers, sales, on='customer_id', how='inner')
left_merge_sales = pd.merge(customers, sales, on='customer_id', how='left')

print("Inner Merge (customers with sales):\n", inner_merge_sales, "\n")
print("Left Merge (all customers, sales may be NaN):\n", left_merge_sales, "\n")

# --- Aggregate Sales Statistics ---
print("Total Sales:", inner_merge_sales['amount'].sum())
print("Average Sales:", inner_merge_sales['amount'].mean())
print("Max Sale:", inner_merge_sales['amount'].max())
print("Sales Count:", inner_merge_sales['amount'].count())

# --- Per Customer Statistics ---
customer_sales = inner_merge_sales.groupby('customer_name')['amount'].sum()
print("\nSales per Customer:\n", customer_sales)

# --- Group by Region ---
region_sales = inner_merge_sales.groupby('region')['amount'].agg(['sum', 'mean', 'max', 'count'])
print("\nSales by Region:\n", region_sales)
```
